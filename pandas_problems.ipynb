## Find Nth Salary
Find the nth highest salary from the Employee table. If there is no nth highest salary, return null.

```python
import pandas as pd
```

```python
data = [[1, 100], [2, 200], [3, 300]]
employee = pd.DataFrame(data, columns=['Id', 'salary']).astype({'Id':'Int64', 'salary':'Int64'})
```

```python
employee
```

### Solution

```python
def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:

    #Create a list of uniuqe salary records. Drop duplicate salary
    unique_salary = employee['salary'].drop_duplicates()

    #Sort unique_salary in descending order and get 2nd highest salary
    nth_highest = unique_salary.nlargest(N).iloc[-1] if len(unique_salary) >=N and N >=1 else None

    #Create a dataframe returning SecondHighestSalary value
    if nth_highest is None:
        return pd.DataFrame({f'getNthHighestSalary({N})' : [None]})
    else:
        return pd.DataFrame({f'getNthHighestSalary({N})' : [nth_highest]})

```

```python
result = nth_highest_salary(employee, -1)
print(result)

result = nth_highest_salary(employee, 2)
print(result)
```

## Consecutive Numbers
Find all numbers that appear at least three times consecutively.
Return the result table in any order.

```python
##  Test case DF #1
data = [[1, 1], [2, 1], [3, 1], [4, 2], [5, 1], [6, 2], [7, 2]]
logs = pd.DataFrame(data, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})

##  Test case DF #2
data2 = [[1, 1]]
logs2 = pd.DataFrame(data2, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
logs2

##  Test case DF #3
data3 = [[1, 3], [2, 3], [3, 3], [4, 3]]
logs3 = pd.DataFrame(data3, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
logs3

##  Test case DF #4
data4 = [[1, 1], [2, 1], [4, 1], [5, 1],[6, 2], [7, 1]]
logs4 = pd.DataFrame(data4, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
logs4
```

```python
import pandas as pd

def consecutive_numbers(logs: pd.DataFrame) -> pd.DataFrame:

    #Let us sort the list first
    logs.sort_values(by=["id"], inplace=True)

    #create a list of id and num values
    id_list = logs["id"].values
    num_list = logs["num"].values

    # Initialize an empty DataFrame for results
    result = pd.DataFrame(columns=['ConsecutiveNums'])

    # Check for edge cases: empty or single-row DataFrame
    if len(num_list) < 3:
        return result  # Return an empty DataFrame if no consecutive triplets are possible

    #Now, let us loop over the lists and compare if 3 consecutive num values are the same
    for i in range(len(id_list)-2):
        if num_list[i] == num_list[i+1] == num_list[i+2] and id_list[i] + 1 == id_list[i+1] and id_list[i] + 2 == id_list[i+2]:
            print(num_list[i])
            #return the value as a Dataframe
            result = result._append({'ConsecutiveNums': num_list[i]}, ignore_index=True)
            result = pd.DataFrame(result['ConsecutiveNums'].unique(), columns=['ConsecutiveNums'])
    return result
```
